```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('big basket.xlsx',header=None)
```

```python
df
```

```python
transaction=[]
for i in range(7219):
    transaction.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
transaction
```

# apriori training on data set

```python
from apyori import apriori
basket=apriori(transactions=transaction,min_support=0.002,min_confidence=0.002,min_lift=3,min_length=2,max_length=2)
```

```python
res=list(basket)
```

```python
res
```

# visualizing the results

```python
def inspect(res):
    p1=[tuple(resu[2][0][0])[0] for resu in res]
    p2=[tuple(resu[2][0][1])[0] for resu in res]
    con=[resu[2][0][2] for resu in res]
    sup=[resu[1] for resu in res]
    lif=[resu[2][0][3] for resu in res]
    return list(zip(p1,p2,sup,con,lif))
d_f=pd.DataFrame(inspect(res),columns=['products 1','product 2','support','confidense','lift'])
```
